```python
from IPython.utils.text import Path
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import drive
import os
# Importing required Packages
drive.mount("/content/drive")     # Connect Drive to Colab
img = cv2.imread("/content/drive/My Drive/Flower.jpg",0)    # Reading the image from the google drive in grayscale
cv2_imshow(img)  # Displaying the image
kernel = np.zeros((15,15), np.uint8)   # creating a kernel of size 15x15 on images
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)   # Removing the white noices (in general it is dilation followed by erosion)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5)) # Returns a elliptical structural element of the specified shape and size of morphological operations.
image = cv2.erode(closing,kernel)
boundaries = closing-image
cv2_imshow(boundaries)
```
